```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, date
from scipy.stats import shapiro
from statsmodels.graphics.gofplots import qqplot
import statsmodels.api as sm
from scipy.optimize import minimize
import warnings

np.random.seed(0)
```

```python
class Stock_Data_Fetcher():
    def __init__(self, stocks, start_date, end_date=None):
        self.stocks = stocks
        self.start_date = start_date
        self.end_date = end_date if end_date is not None else date.today().strftime('%Y-%m-%d')
        self.data = self.fetch_stock_data()

    def fetch_stock_data(self):
        data = yf.download(self.stocks, start=self.start_date, end=self.end_date)['Adj Close']
        return data
    
    def calculate_log_returns(self):
        log_returns = self.data.pct_change().dropna()
        self.log_returns = log_returns.apply(lambda x: np.log(1 + x))
        
    def draw_date_coloured_scatterplot(self, pair):
   
        # Create a yellow-to-red colourmap where yellow indicates
        # early dates and red indicates later dates
        plen = len(self.data)
        a, b = pair[0], pair[1]
        colour_map = plt.cm.get_cmap('YlOrRd')    
        colours = np.linspace(0.1, 1, plen)

        # Create the scatterplot object
        scatterplot = plt.scatter(
            self.data[a], self.data[b], 
            s=30, c=colours, cmap=colour_map, 
            edgecolor='k', alpha=0.8
        )

        # Add a colour bar for the date colouring and set the 
        # corresponding axis tick labels to equal string-formatted dates
        colourbar = plt.colorbar(scatterplot)
        colourbar.ax.set_yticklabels(
            [str(p.date()) for p in self.data[::plen//9].index]
        )
        plt.grid()
        plt.xlabel(self.data.columns[0])
        plt.ylabel(self.data.columns[1])
        plt.show()
    
```

```python
stocks = ['SGOV', 'SHV', 'SHY', 'IEI', 'IEF', 'TLH', 'TLT', 'GOVZ']
# stocks = ['AAPL', 'AMZN', 'GOOG', 'MSFT', 'TSLA', 'JNJ', 'MA', 'KO']
start_date = '2021-01-01'
end_date = None


fetcher = Stock_Data_Fetcher(stocks, start_date, end_date)
fetcher.calculate_log_returns()
# df = fetcher.log_returns
df = fetcher.data
```

```python
split_date = pd.to_datetime('2023-07-01')
PCA_lb = 365 # 1 year
PCA_start = split_date - pd.DateOffset(days=PCA_lb)
df_fit_PCA = df[(df.index <= split_date) & (df.index >= PCA_start)]
df_test = df[(df.index > split_date)]
```

```python
# https://stats.stackexchange.com/questions/229092/how-to-reverse-pca-and-reconstruct-original-variables-from-several-principal-com
```

```python
X = ((df_fit_PCA - df_fit_PCA.mean()) / df_fit_PCA.std()).to_numpy()
cov = np.cov(X.T)
U, s, Vh = np.linalg.svd(cov)
e_vals = s ** 2
explained_variances = e_vals / np.sum(e_vals)
e_vecs = Vh.T # e_vecs are columns
e_vecs_red = e_vecs[:, :3]
T = X @ e_vecs_red # component scores "how much of each data point alings with the (unit) weight vectors (eigenvectors of Cov)"
# Component scores 
X_recon = T @ e_vecs_red.T * df_fit_PCA.std().to_numpy() + df_fit_PCA.mean().to_numpy() # Now turn the scalar scores for each (unit) weight vector into a vector and unstandardize. (reconstruct)  
```

```python
plt.bar(range(1, len(explained_variances) + 1), explained_variances)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance')
plt.show()
```

```python
df_fit_recon = pd.DataFrame(X_recon, index=df_fit_PCA.index, columns=df_fit_PCA.columns)
df_fit_spread = df_fit_PCA - df_fit_recon
```

```python
fig, ax = plt.subplots(nrows=len(df_fit_PCA.columns), figsize=(8,16))
for i, col in enumerate(df_fit_PCA.columns):
    ax[i].plot(df_fit_recon[col], label='PCA')
    ax[i].plot(df_fit_PCA[col], label=f'{col}')
    ax[i].legend()
    ax[i].grid()
```

```python
# Plot the spread
fig, ax = plt.subplots(nrows=len(df_fit_PCA.columns), figsize=(8,16))
for i, col in enumerate(df_fit_spread.columns):
    ax[i].plot((df_fit_spread[col] - df_fit_spread[col].mean())/df_fit_spread[col].std(), label=f"{col} PCA spread")
    print(f"{col} mean {df_fit_spread[col].mean():.2f}, std {df_fit_spread[col].std():.2f}")
    ax[i].legend()
    ax[i].grid()
```

```python
# See if these are stationary 
AR1_params = {}
for i, col in enumerate(df_fit_spread.columns):
    model = sm.tsa.ARIMA(df_fit_spread[col], order=(1, 0, 0))
    results = model.fit()
    AR1_params[col] = results.arparams[0]
```
